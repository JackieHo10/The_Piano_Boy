```python
import matplotlib.pyplot as plt


epoch = []
loss = []
i = 1
with open('log.txt','r') as file:
    for line in file:
        line = line.strip()
        if(line.find('loss') != -1):
            line = line.split()
            print(line)
#             print(line[7])
#             line[5] = line[5].split(',')[0]
            print(line[7])
            epoch.append(i)
            i += 1
            loss.append(float(line[7]))
# for i in iteration:
#     print i
# for i in loss:
#     print i


_, ax1 = plt.subplots()
ax1.plot(epoch, loss)
ax1.set_xlabel('epoch')
ax1.set_ylabel('train loss')

plt.show()
```
